# The Three Doors Problem

### importing some modules

```python
%matplotlib inline
import matplotlib.pyplot as plt
import random
```

### Define the function that simulates the game

```python
def threeDoors(n, s):
    l = ['door A', 'door B', 'door C']
    percentage = []
    wins = 0.0

    for i in range(1, n + 1):
        winningDoor = random.choice(l)
        playerPick01 = random.choice(l)
        openedDoor = random.choice([j for j in l if j != winningDoor and j != playerPick01])

        if s:
            playerPick02 = [j for j in l if j != playerPick01 and j != openedDoor][0]
            if playerPick02 == winningDoor:
                wins += 1
        else:
            if playerPick01 == winningDoor:
                wins += 1

        percentage.append((wins / i) * 100)

    return percentage
```

### Start the simulation
First the game is played 10000 times with the player switching doors each time.
Then the game is played another 10000 times with the player not switching. The results are stored in two lists.

```python
switched = threeDoors(10000, True)
notSwitched = threeDoors(10000, False)
```

### Plotting the result in a graph

```python
fig = plt.figure(figsize=(20, 10))
plt.style.use('ggplot')
plt.plot(switched, label='player switched', color = 'green')
plt.plot(notSwitched, label='player did not switch', color = 'red')
plt.legend()
plt.ylim(0, 100)
plt.grid(False)
plt.title('Three Door Problem')
plt.ylabel('percentage games won by player')
plt.xlabel('games played')

plt.plot([0, 10000], [33, 33], color = 'k', linestyle = '--')
plt.plot([0, 10000], [66, 66], color = 'k', linestyle = '--')

plt.show()
```
